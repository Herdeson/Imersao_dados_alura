**Importação dos Dados**

```python
import pandas as pd
url_dados = 'https://github.com/alura-cursos/imersaodados3/blob/main/dados/dados_experimentos.zip?raw=true'

dados = pd.read_csv(url_dados, compression='zip')
dados
```

**DESAFIO 01: Investigar por que a classe tratamento é tão desbalanceada?**

```python
dados['tratamento'].value_counts(normalize=True)
```

```python
dados.groupby(['tratamento', 'droga']).count()
```

Como podemos observar a classe com dogras é maior por conta de uma maior combinação possível variáveis a serem testadas. Cada droga é variada na dosagem aplicada nos períodos de observações e nos efeitos das colunas g-0 até g-99.

**DESAFIO 02: listar 05 últimas linhas da tabela**

Por padrão as funções tail e head possuem como parâmetro o total de linhas e por padrão as duas funções apresentam o valor 5.

```python
dados.tail()
```

**DESAFIO 03: Proporção das classes tratamento**

 Parâmetro normalize da função value_counts retorna a proporção entre as frequências.

```python
dados['tratamento'].value_counts(normalize=True)
```

**Desafio 04: Quantos tipos de drogas foram investigados.**

```python
dados['droga'].describe()
```

Como observado acima, o total de drogas investigadas foram 3289. A droga que aparece com mais frequência é a droga: *cacb2b860*. Ela apresenta uma frequência de 1866 vezes aplicada.


**Desafio 05:  Procurar na documentação o método query(pandas).**

Por motivos de o interpretador ter problemas com hífen no meio das colunas, foi necessário utilizar  o apóstrofo para funcionamento correto da função. 

```python
filtrados = dados.query('`g-0` > 0')
filtrados.head()
```

**Desafio 06: Renomear as colunas tirando o hífen.**

```python
dados.columns = dados.columns.str.replace('-','')
dados.columns
```

**Desafio 07: Deixar os gráficos bonitões. (Matplotlib.pyplot)**

```python
import matplotlib.pyplot as plt
plt.title("Total de análises por tempo de observação")
dados['tempo'].value_counts().plot.bar()
plt.xlabel("Tempo de observação (h)")
plt.ylabel("Total")
#plt.figure(figsize=(10,4))
plt.show()
```
